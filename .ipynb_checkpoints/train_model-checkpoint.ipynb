Now visualize `house_value` vs every variable to observe if there are any hideous outliers which need to be removed.

```python
# Visualize house_value vs every variable to observe if there is any hideous outliers
f, axarr = plt.subplots(3, 2, figsize=(15,15))
k = 0
for i in range(3):
    for j in range(2):
        if k != 5:
            axarr[i, j].scatter(x = housing.iloc[:, k], y = housing['house_value'])
            axarr[i, j].set_ylabel('house_value', fontsize=13)
            axarr[i, j].set_xlabel(housing.columns[k], fontsize=13)
            k += 1

sns.distplot(housing['house_value'] , fit=norm);

# Fit house value by normal distribution
(mu, sigma) = norm.fit(housing['house_value'])

# Plot the distribution
axarr[2, 1].legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
axarr[2, 1].set_ylabel('Frequency')
axarr[2, 1].set_title('House value distribution')
plt.setp(axarr, yticks=[])
plt.tight_layout()

f, axarr = plt.subplots(3, 2, figsize=(15,15))
k = 0
for i in range(3):
    for j in range(2):
        if k != 5:
            sns.distplot(housing.iloc[:, k] , fit=norm, ax=axarr[i, j]);
            (mu, sigma) = norm.fit(housing[housing.columns[k]])
            axarr[i, j].legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
            axarr[i, j].set_ylabel('Frequency')
            axarr[i, j].set_title(housing.columns[k] + ' distribution')
            k += 1
```

One observation:

* No extremely hedious outliders were observed. So no need to remove any data point.

```python
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import KFold
from sklearn.linear_model import BayesianRidge
import pickle

class regression_model(object):
    def __init__(self, reg_model, params=None, to_scale=True):
#         params['random_state'] = seed
        self.reg_model = reg_model(**params)
        self.to_scale = to_scale

    def train(self, X, y):
        if self.to_scale:
            self.reg_model.fit(transformer.transform(X), y)
        else:
            self.reg_model.fit(X, y)        

    def predict(self, X):
        if self.to_scale:
            return self.reg_model.predict(transformer.transform(X))
        else:
            return self.reg_model.predict(X)
    
    def fit(self, X, y):
        if self.to_scale:
            return self.reg_model.fit(transformer.transform(X), y)
        else:
            return self.reg_model.fit(X, y)
```

### Now calculate the mean squared error in a 5-fold cross validation

```python
housing = pd.read_csv('./housing.csv')

predictors = ["crime_rate", "avg_number_of_rooms", "distance_to_employment_centers",
              "property_tax_rate", "pupil_teacher_ratio"]
X = housing[predictors]
y = housing['house_value']

transformer = MinMaxScaler(feature_range=(0, 1)).fit(X)
pickle.dump(transformer, open( "transformer.p", "wb" ))

br = {"alpha_1": 1e-6}
methods = ['br']


pred = pd.DataFrame(np.empty((X.shape[0], len(methods))), columns=methods)
scores = np.zeros(5)
i = 0
kf = KFold(n_splits=5, shuffle=True, random_state=1354)
for train, test in kf.split(X):
    X_train = X.iloc[train, :]
    y_train = y.iloc[train]
    X_test = X.iloc[test, :]
    y_test = y.iloc[test]
    
    name_method = 'br'
    model = regression_model(reg_model=BayesianRidge, params=eval(name_method))
    model.fit(X_train, y_train)
    pred[name_method][test] = model.predict(X_test)
    
    scores[i] = mean_squared_error(y_test, model.predict(X_test))   
    i += 1
    
print(scores.mean())
```
